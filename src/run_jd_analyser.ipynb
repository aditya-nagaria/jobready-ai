# JobReady AI — JD Analyser
> *Turning a life-ending moment into a life-changing phase.*

**What this notebook does:**
1. Takes your profile and a job description as input
2. Sends it to an LLM (Gemini or OpenAI)
3. Returns a fit score, gap analysis, ATS keywords and recommendations
4. Saves a formatted PDF report

**Modes:**
- `mock` → runs without an API key (tests the full pipeline with realistic fake data)
- `real` → calls the actual Gemini or OpenAI API (paste your key in `config.py`)

## Switching to real mode

When you are ready to use a real API key:

1. Open `config.py`
2. Set `LLM_MODE = 'real'`
3. Set `LLM_PROVIDER = 'gemini'` (or `'openai'`)
4. Paste your API key in `GEMINI_API_KEY` or `OPENAI_API_KEY`
5. Uncomment the relevant `pip install` in Step 1
6. Re-run from Step 1
